```python

class book:
    def __init__(self,identifier,title,author):
        self.identifier = identifier
        self.title = title
        self.author = author

    def __str__(self):
            return f"Book(ID: {self.identifier}, Title: '{self.title}', Author: '{self.author}')"

class Library:
    def __init__(self):
        self.catalogue={}

    def addBook(self, book):
        if book.identifier in self.catalogue:
            print(f"Buku {book.identifier} sudah ada di katalog.")
        else:
          self.catalogue[book.identifier]=book
          print(f"Buku: {book.title} berhasil ditambahkan")
          
    def search_by_title(self, title):
        results = {}
        for book in self.catalogue.values():
          if title.lower() == book.title.lower():
            results=book
          else:
            print("Buku tidak ditemukan")
        print(results)
    
    def search_by_author(self, author):
        results = {}
        for book in self.catalogue.values():
          if author.lower() == book.author.lower():
            results=book
          else:
            print("Buku tidak ditemukan")
        print(results)
    
    def remove_book(self, identifier):        
        if identifier not in self.catalogue:            
            print(f"Buku {identifier} tidak ditemukan.")
        else:
          removed_book = self.catalogue.pop(identifier)       
          print(f"Buku '{removed_book.title}' berhasil dihapus.")

    def showBook(self):
        if not self.catalogue:
            print("Katalog tidak tersedia")
        else:            
            print("Daftar Buku:")            
            for book in self.catalogue.values():                
                print(book)

if __name__ == "__main__":

    lib = Library()

    a=True
    while(a==True):
        print()

        print("===================")
        print("Program Management Buku")
        print("===================")
        print("MENU:")
        print("1. Tambah Buku")
        print("2. Buang Buku")
        print("3. Cari Buku berdasar judul")
        print("4. Cari Buku berdasar pengarang")
        print("5. Tampilkan seluruh buku")
        print("Exit")

        print("Pilih menu yang di diinginkan:")
        inputdata =str(input())

        print()

        if inputdata == '1':
            identifier=input("Masukkan ID Buku: ")
            title=input("Masukkan Nama Buku: ")
            author=input("Masukkan Pengarang Buku: ")
            databook=book(identifier,title,author)
            lib.addBook(databook)
        
        elif inputdata == '2':
            id=input("Masukkan ID buku yang ingin dihapus: ")
            lib.remove_book(id)
        
        elif inputdata == '3':
            judul=input("Masukkan judul buku :")
            lib.search_by_title(judul)
        
        elif inputdata == '4':
            auth=input("Masukkan nama pengarang buku :")
            lib.search_by_author(auth)

        elif inputdata == '5':
            lib.showBook()

        elif inputdata=='exit':
            print('Terima Kasih!')
            a=False
        else:
            print('input tidak valid')
```
